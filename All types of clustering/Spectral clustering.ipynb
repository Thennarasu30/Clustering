```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt

list1 = []  # To store WCSS values

for i in range(2, 11):  # Loop through different numbers of clusters
    spectral = SpectralClustering(n_clusters=i, random_state=42)
    spectral.fit(x)
    labels = spectral.labels_
    
    # Calculate WCSS manually
    wcss = 0
    for j in range(i):
        cluster_points = x[labels == j]
        centroid = np.mean(cluster_points, axis=0)
        wcss += np.sum((cluster_points - centroid) ** 2)
    
    list1.append(wcss)  # Append WCSS to the list

# Plot the Elbow graph
plt.plot(range(2, 11), list1)
plt.title('The spectral clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
