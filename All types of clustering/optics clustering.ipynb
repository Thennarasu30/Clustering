```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
op
import matplotlib.pyplot as plt

list1 = []  # To store number of clusters
eps_values = np.arange(0.1, 1.1, 0.1)  # Loop through different epsilon values

for eps in eps_values:
    # Perform OPTICS clustering
    optics = OPTICS(min_samples=10, xi=.05, min_cluster_size=10).fit(x)
    labels = optics.labels_
    
    # Calculate the number of clusters
    num_clusters = len(np.unique(labels))
    
    list1.append(num_clusters)  # Append number of clusters to the list

# Plot the graph
plt.plot(eps_values, list1)
plt.title('OPTICS Clustering')
plt.xlabel('Epsilon Values')
plt.ylabel('Number of Clusters')
plt.show()

```

```python
from sklearn.cluster import OPTICS
op=OPTICS(min_samples=2)
y_op=op.fit_predict(x)
```
